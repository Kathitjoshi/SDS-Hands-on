```python
import numpy as np
from math import sqrt
from matplotlib import pyplot
from scipy.stats import norm, zscore
z=1.96
n=64
moe=[]
std_dev=range(21)
```

```python
for i in std_dev:
    moe.append(z*(i/sqrt(n)))
print(moe)
pyplot.plot(std_dev,moe)
pyplot.xlabel('std_dev')
pyplot.ylabel('moe')
```
